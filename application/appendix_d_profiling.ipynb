```python
import pandas as pd
```

```python
paths = ["/Workspace/Users/armando.n90@gmail.com/users_case/files/20130208_000000/report_7.txt", 
         "/Workspace/Users/armando.n90@gmail.com/users_case/files/20130208_000000/report_8.txt",
         "/Workspace/Users/armando.n90@gmail.com/users_case/files/20130214_000000/report_9.txt"]

df_raw = []
for path in paths:
    df = pd.read_csv(path)
    print(df.columns)

    if 'jk' in df.columns:
        df = df.rename(columns={'jk': 'jyv'})

    if 'fgh' in df.columns:
        df = df.rename(columns={'fgh': 'jyv'})

    print('Total of rows in this batch: ' + str(df.shape[0]))
    print('Total un unique emails: ' + str(df['email'].unique().shape[0]))
    df_raw.append(df)
    print()

df_raw = pd.concat(df_raw)
print(df_raw.shape)
display(df_raw.head(5))
```

```python
print(df_raw['email'].unique().shape)
```

```python
conteo = df_raw.groupby(by='email', as_index=False).agg({'Fecha envio': 'count'})
conteo = conteo.sort_values(by='Fecha envio', ascending=False)
conteo = conteo.rename(columns={'Fecha envio': 'conteo'})

print(conteo['conteo'].value_counts())
```
